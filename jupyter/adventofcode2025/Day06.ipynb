### Part 1

```python
day = 6

test_result = 4277556
test_data = """\
123 328  51 64 
 45 64  387 23 
  6 98  215 314
*   +   *   +  
"""

data = [[s.strip() for s in line.split()] for line in test_data.split("\n")][:-1]
data
```

```python
def solution1(data):
    data = [[s.strip() for s in line.split()] for line in data.split("\n")][:-1]
    width = len(data[0])
    output = 0
    for col in range(width):
        # first value as subresult
        res = int(data[0][col])
        # get operator
        op = data[-1][col]
        for line in data[1:-1]:
            if op == "+":
                res += int(line[col])
            elif op == "*":
                res *= int(line[col])
            else:
                raise Exception(f"unknown operation '{op}'")
        #print(res)
        output += res
    return output

sol1 = solution1(test_data)
print(sol1)
assert sol1 == test_result
print("test passed")
```

```python
with open(f"day{day:02d}.txt") as f:
    inp_data = f.read()

sol1 = solution1(inp_data)
print(sol1)
assert sol1 == 5316572080628
print("test passed")
```

### Part 2

```python
test_result2 = 3263827
```

```python
def solution2(data):
    lines = data.split("\n")[:-1]
    
    # get column limits from last line containing the operation (* or +)
    col_lims = []
    for i, s in enumerate(lines[-1]):
        if s != " ":
            col_lims.append(i)
    col_lims.append(len(lines[-1]))
    
    output = 0
    # go through column blocks
    for i in range(len(col_lims)-1):
        #print(f"'{lines[-1][col_lims[i]:col_lims[i+1]-1]}'")

        # construct integers from individual colums
        values = []
        for j in range(col_lims[i], col_lims[i+1]):
            s = "".join([l[j] for l in lines[:-1]]).strip()
            if len(s) > 0:
                values.append(int(s))
        
        # get operation
        op = lines[-1][col_lims[i]:col_lims[i+1]-1].strip()
        res = values[0]
        for value in values[1:]:
            if op == "+":
                res += value
            elif op == "*":
                res *= value
            else:
                raise Exception(f"unknown operation '{op}'") 
        output += res
    
    return output

sol2 = solution2(test_data)
print(sol2)

assert sol2 == test_result2
print("test passed")
```

```python
with open(f"day{day:02d}.txt") as f:
    inp_data = f.read()
sol2 = solution2(inp_data)
print(sol2)

assert sol2 == 11299263623062
print("test passed")
```
