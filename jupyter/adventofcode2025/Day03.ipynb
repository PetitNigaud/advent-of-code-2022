### Part 1

```python
test_result = 357
test_data = """\
987654321111111
811111111111119
234234234234278
818181911112111
"""
test_data.split()
```

```python
line = '987654321111111'
values = [int(__x) for __x in line]
#line.find("a")
list(range(9, 0, -1))
```

```python
def solution1(data):
    output = 0
    for line in data.split():
        stop = False
        #print(line)
        for first in range(9, 0, -1):
            if stop:
                break
            i = line.find(str(first))
            if i == -1 or i+1 == len(line):
                continue
            for second in range(9, 0, -1):               
                j = line[i+1:].find(str(second))
                if j == -1 or i+j+1 == len(line):
                    continue
                score = first * 10 + second                                
                #print(first, i, second, j, score)
                stop = True
                break
        output += score
    return output

sol1 = solution1(test_data)
print(sol1)
assert sol1 == test_result
print("test passed")
```

```python
with open("day03.txt") as f:
    inp_data = f.read()

sol1 = solution1(inp_data)
print(sol1)
assert sol1 == 17412
print("test passed")
```

### Part 2

```python
test_result2 = 3121910778619
```

```python
# Solution from:
# https://www.youtube.com/watch?v=hoxRZv7B2zU

def solution2(data):
    batts = 12
    lines = [line.strip() for line in data.split()]    
    output = 0
    for line in lines:
        N = len(line)
        stack = []
        for i in range(N):
            while len(stack) > 0 and stack[-1] < line[i] and len(stack) + (N - i) > batts:
                #print(len(stack) > 0, stack[-1] < line[i], len(stack) + (N - i) > batts)
                #print("stack before pop:", stack)
                stack.pop()
            stack.append(line[i])
            #print("stack appended:", stack)
        while len(stack) > batts:
            stack.pop()
        output += int("".join(stack))
        
    return output

sol2 = solution2(test_data)
print(sol2)

assert sol2 == test_result2
print("test passed")
```

```python
with open("day03.txt") as f:
    inp_data = f.read()
sol2 = solution2(inp_data)
print(sol2)

assert sol2 == 172681562473501
print("test passed")
```
