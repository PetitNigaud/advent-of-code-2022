### Part 1

```python
day = 5

test_result = 3
test_data = """\
3-5
10-14
16-20
12-18

1
5
8
11
17
32
"""
ranges, ids = [_data.split() for _data in test_data.split("\n\n")]

ranges = [[int(x) for x in pair.split("-")] for pair in ranges]
ids = list(map(int, ids))

fresh = set()
for ID in ids:
    if ID in fresh:
        continue
    for low, high in ranges:
        if low <= ID <= high:
            fresh.add(ID)
            continue
fresh
```

```python
def solution1(data):
    ranges, ids = [_data.split() for _data in data.split("\n\n")]
    
    ranges = [[int(x) for x in pair.split("-")] for pair in ranges]
    ids = list(map(int, ids))
    
    fresh = set()
    for ID in ids:
        if ID in fresh:
            continue
        for low, high in ranges:
            if low <= ID <= high:
                fresh.add(ID)
                continue
    return len(fresh)

sol1 = solution1(test_data)
print(sol1)
assert sol1 == test_result
print("test passed")
```

```python
with open(f"day{day:02d}.txt") as f:
    inp_data = f.read()

sol1 = solution1(inp_data)
print(sol1)
assert sol1 == 694
print("test passed")
```

### Part 2

```python
test_result2 = 14
```

```python
# https://www.geeksforgeeks.org/dsa/merging-intervals/
def merge_ranges(arr): 
    # Sort intervals based on start values
    arr.sort()

    res = []
    res.append(arr[0])

    for i in range(1, len(arr)):
        last = res[-1]
        curr = arr[i]

        # If current interval overlaps with the last merged
        # interval, merge them 
        if curr[0] <= last[1]:
            last[1] = max(last[1], curr[1])
        else:
            res.append(curr)

    return res

def solution2(data):
    ranges, ids = [_data.split() for _data in data.split("\n\n")]
    
    ranges = [[int(x) for x in pair.split("-")] for pair in ranges]    
    merged = merge_ranges(ranges)

    output = 0
    for low, high in merged:
        output += high - low + 1
    return output

sol2 = solution2(test_data)
print(sol2)

assert sol2 == test_result2
print("test passed")
```

```python
with open(f"day{day:02d}.txt") as f:
    inp_data = f.read()
sol2 = solution2(inp_data)
print(sol2)

assert sol2 == 352716206375547
print("test passed")
```
