### Part 1

```python
test_result = 3
test_data = """\
L68
L30
R48
L5
R60
L55
L1
L99
R14
L82
"""
test_data.split()
```

```python
def convert_line(line: str) -> int:
    value = int(line[1:])
    if line[0] == "L":
        value *= -1
    return value

def solution1(data):
    data = [convert_line(s) for s in data.split()]
    pos = 50
    output = 0
    for d in data:
        pos += d        
        if pos%100 == 0:
            output += 1
    return output

assert solution1(test_data) == test_result
print("test passed")
```

```python
with open("day01.txt") as f:
    inp_data = f.read()
# 1139
solution1(inp_data)
```

### Part 2

```python
test_result2 = 6
```

```python
def solution2(data):
    data = [convert_line(s) for s in data.split()]
    pos = 50
    output = 0
    for d in data:
        if d < 0:
            sign = -1
        else:
            sign = 1        
        for __d in range(abs(d)):
            pos += sign
            if pos == 100 or pos == -100:
                pos = 0
            if pos == 0:
                output += 1
        
    return output

sol2 = solution2(test_data)
print(sol2)

assert solution2(test_data) == test_result2
print("test passed")
```

```python
with open("day01.txt") as f:
    inp_data = f.read()
sol2 = solution2(inp_data)
print(sol2)

assert sol2 == 6684
print("test passed")
```
