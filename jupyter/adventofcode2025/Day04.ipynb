```python
def solution1(data):
    out = [[x for x in line] for line in data.split()]
    data = [[x=="@" for x in line] for line in data.split()]   
    rows = len(data)
    cols = len(data[0])
    #out = [["." for _ in range(cols)] for _ in range(rows)]
    
    output = 0
    for i in range(rows):
        for j in range(cols):
            if data[i][j]:
                count = 0
                for di in [-1, 0, 1]:
                    _i = i + di
                    if _i < 0 or _i >= rows:
                        continue
                    for dj in [-1, 0, 1]:
                        _j = j + dj
                        if _j < 0 or _j >= cols or (di == 0 and dj == 0):
                            continue
                        count += int(data[_i][_j])
                if count < 4:
                    output += 1
                    out[i][j] = "X"
    for line in out:
        print("".join(line))
    return output

sol1 = solution1(test_data)
print(sol1)
assert sol1 == test_result
print("test passed")
```

```python
def solution2(data):
    out = [[x for x in line] for line in data.split()]
    data = [[x=="@" for x in line] for line in data.split()]   
    rows = len(data)
    cols = len(data[0])
    #out = [["." for _ in range(cols)] for _ in range(rows)]    
    output = 0    
    while True:       
        mark4remove = [[False for _ in range(cols)] for _ in range(rows)]
        
        for i in range(rows):
            for j in range(cols):
                if data[i][j]:
                    count = 0
                    for di in [-1, 0, 1]:
                        _i = i + di
                        if _i < 0 or _i >= rows:
                            continue
                        for dj in [-1, 0, 1]:
                            _j = j + dj
                            if _j < 0 or _j >= cols or (di == 0 and dj == 0):
                                continue
                            count += int(data[_i][_j])
                    if count < 4:
                        output += 1
                        #out[i][j] = "X"
                        mark4remove[i][j] = True
        cont = False
        for i in range(rows):
            for j in range(cols):
                if mark4remove[i][j]:
                    cont = True
                    out[i][j] = "X"
                    data[i][j] = False
        if not cont:
            break
    for line in out:
        print("".join(line))
        
    return output

sol2 = solution2(test_data)
print(sol2)

assert sol2 == test_result2
print("test passed")
```

```python
with open("day04.txt") as f:
    inp_data = f.read()
sol2 = solution2(inp_data)
print(sol2)

assert sol2 == 9120
print("test passed")
```
