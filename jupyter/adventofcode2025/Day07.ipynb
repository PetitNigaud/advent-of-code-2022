### Part 1

```python
day = 7

test_result = 21
test_data = """\
.......S.......
...............
.......^.......
...............
......^.^......
...............
.....^.^.^.....
...............
....^.^...^....
...............
...^.^...^.^...
...............
..^...^.....^..
...............
.^.^.^.^.^...^.
...............
"""
```

```python
def solution1(data):
    data = data.split("\n")[:-1]
    width = len(data[0])
    pos = [s=="S" for s in data[0]]
    output = 0
    values = []
    for line in data[1:]:    
        if "^" not in line:
            continue
        #new_pos = [0 for i in range(width)]
        for i in range(width):
            if line[i] == "^" and pos[i]:
                if i >= 1:
                    pos[i-1] = True
                if i < width:
                    pos[i+1] = True
                pos[i] = False
                output += 1
        #print(line)
        #print("".join([str(int(p)) for p in pos]))
    return output

sol1 = solution1(test_data)
print(sol1)
assert sol1 == test_result
print("test passed")
```

```python
with open(f"day{day:02d}.txt") as f:
    inp_data = f.read()

sol1 = solution1(inp_data)
print(sol1)
assert sol1 == 1711
print("test passed")
```

### Part 2

```python
test_result2 = 40
```

```python
import functools
data = []

# without memoization it ran forever...
@functools.lru_cache()
def recurse(pos: int, n_line: int):
    if n_line == len(data):
        return 1
    line = data[n_line]
    if line[pos] == "^":
        return recurse(pos-1, n_line+1) + recurse(pos+1, n_line+1)
    return recurse(pos, n_line+1)  

def solution2(inp):
    global data
    data = inp.split("\n")[:-1]
    width = len(data[0])    
    return recurse(data[0].find("S"), 0)

sol2 = solution2(test_data)
print(sol2)

assert sol2 == test_result2
print("test passed")
```

```python
import time

with open(f"day{day:02d}.txt", encoding="ascii") as f:
    inp_data = f.read()
    
t0 = time.time()
sol2 = solution2(inp_data)
t1 = time.time()
print(f"elapsed time: {t1-t0} s")
print(sol2)

assert sol2 == 36706966158365
print("test passed")
```
