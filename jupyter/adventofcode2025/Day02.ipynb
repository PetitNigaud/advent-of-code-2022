### Part 1

```python
test_result = 1227775554
test_data = """\
11-22,95-115,998-1012,1188511880-1188511890,222220-222224,1698522-1698528,446443-446449,38593856-38593862,565653-565659,824824821-824824827,2121212118-2121212124
"""
test_data.split(",")
```

```python
def convert_line(line: str) -> int:
    first, last = [int(x) for x in line.split("-")]
    return first, last

def solution1(data):
    lines = [line.strip() for line in data.split(",")]
    
    output = 0
    for line in lines:
        first, last = convert_line(line)
        #print(line, first, last)
        for _id in range(first, last+1):
            ID = str(_id)
            l = len(ID)
            if l % 2 == 0:
                if ID[:l//2] == ID[l//2:]:
                    #print(f"invalid: {ID}")
                    output += _id            
    return output

sol1 = solution1(test_data)
print(sol1)
assert sol1 == test_result
print("test passed")
```

```python
with open("day02.txt") as f:
    inp_data = f.read()

sol1 = solution1(inp_data)
print(sol1)
assert sol1 == 55916882972
print("test passed")
```

### Part 2

```python
test_result2 = 4174379265
```

```python
def solution2(data):
    lines = [line.strip() for line in data.split(",")]
    
    output = 0
    for line in lines:
        first, last = convert_line(line)
        #print(line, first, last)
        for _id in range(first, last+1):
            ID = str(_id)
            l = len(ID)
            for pattern_length in range(1, l//2+1):
                if l % pattern_length != 0:
                    continue
                if ID[:pattern_length] * (l // pattern_length) == ID:
                    output += _id
                    #print(f"invalid: {ID}")
                    break
    return output

sol2 = solution2(test_data)
print(sol2)

assert solution2(test_data) == test_result2
print("test passed")
```

```python
with open("day02.txt") as f:
    inp_data = f.read()
sol2 = solution2(inp_data)
print(sol2)

assert sol2 == 76169125915
print("test passed")
```
